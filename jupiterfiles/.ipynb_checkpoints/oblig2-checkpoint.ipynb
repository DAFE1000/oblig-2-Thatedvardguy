```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x/4) * np.arctan(x)

def f_der(x):
    return -(np.arctan(x))*(np.exp(-x/4))/4 + np.exp(-x/4)/(1+x**2)

#def f_der(x):
 #   return -(np.arctan(x))/4 + np.exp(-x/4)/(1+x**2)

def f_DoubleDer(x):
    return (4/(x**2 + 1) + ((8*x)/(1+x**2)**2))

x = np.linspace(-5, 10, 50)
plt.plot(x, f(x), color="blue", label="f(x)")
plt.plot(x, f_der(x), color="green", label ="f'(x)")
plt.scatter(2, 2)
#plt.plot(x, f_doubleDer(x))
plt.grid()
```
