Plotter funksjonene

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x/4) * np.arctan(x)

def f_der(x):
    return np.arctan(x)- (4/(1+x**2))

x = np.linspace(-5, 10, 50)
plt.plot(x, f(x), color="blue", label="f(x)")
plt.plot(x, f_der(x), color="green", label ="f'(x)")
plt.grid()
```

```python
# Skript som implementerer Newtons metode
# Bestemmer x_0
x = 2

# Presisjon
pres = 1e-5

# Initerer talet på iterasjonar
iterasjonar = 0

# Lagar ein variabel for "gamal" x-verdi (tilfeldig tal)
x_gml = 100

while abs(x-x_gml) > pres:
    # Kopierer den gamle x
    x_gml = x
    # Iterasjonsskjema
    x = x - f_der(x)/(1/(1+x**2) + ((8*x)/(1+x**2)**2))
    # Tel iterasjonar
    iterasjonar = iterasjonar+1

# Skriv svaret til skjerm
print(f'x = {x:.10f}')

# Skriv talet på iterasjonar til skjerm
print(f'iterasjonar = {iterasjonar}')
```
